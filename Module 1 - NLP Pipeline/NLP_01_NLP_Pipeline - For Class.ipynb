### Extraction from PDF documents

A PDF page is essentially a sequence of drawing instructions describing how to render text, images, and shapes at specific coordinates.

- PDF processing libraries
    - `PyPDF`: Works at the structural level of the PDF (concatenate pages, split, rotate, encryption, metadata). Mainly used for PDF manipulation.
    - `PDFMiner`: Specialized in text extraction and layout reconstruction.
    - PyPDF2: Deprecated library that has now been replaced by **PyPDF**.
    - `PyMuPDF` (fitz): Exposes the graphical structure of the PDF (text blocks, coordinates, images, shapes), enabling advanced extraction and layout-aware processing.

Structural libraries  → PyPDF

Layout-aware extraction → PyMuPDF / PDFMiner

### Contractions : Expanding contractions
- Use Regular expressions
    - don't --> do not
    - isn't --> is not
    - aren't --> are not
    - we're --> we are
    - they're --> they are

````python
test_sentence = "Everything we’re doing now is great. However, we don't want to relax now. And this isn't the time to relax at all."
````

````python
import re # Regular expressions
pattern = r'we[\’\']re'
replacement = 'we are'
expanded_sentence = re.sub(pattern,replacement,test_sentence)
print(expanded_sentence)
````

#### Generalize the contraction expansion

````python
pattern = r'\w[\’\']re'
replacement = ' are'
expanded_sentence = re.sub(pattern,replacement,test_sentence)
print(expanded_sentence)
````

#### Write one regular expression for don't types

````python
pattern = r'\w[\’\']t'
replacement = ' not'
expanded_sentence = re.sub(pattern,replacement,test_sentence)
print(expanded_sentence)
````

Version without underscore are numerical identifiers, while those with underscore are the actual values. \
For example, `token.lemma_` gives the lemmatized form of the token, while `token.lemma` (without the underscore) would give a numerical ID representing the lemma in spaCy's internal vocabulary.

````python
print(token.lemma_)
print(token.lemma)
````

Common API:
``` /
    text      → Original word
    lemma_    → Base form
    pos_      → Grammar category (Part of Speech tag)
    tag_      → Detailed grammar category
    ent_type_ → Named entity (NER)
    dep_      → Role in the sentence, depends on how it relates with the other words (Dependency Parse)
    head      → Word on which it depends (father node)

    vector    → Token embedding
    idx       → Position in the text
    i         → Token index

Example pos_ vs tag_:
``` /
    pos_  = VERB
    tag_  = VBG   (verb, gerund)
    

````python
print('Lemmatization:')
for token in doc[:9]:
    print(f'    {token.text} --> {token.lemma_}')
    
print('\n')
print('Grammar Category (POS tag):')
for token in doc[:9]:
    print(f'    {token.text} --> {token.pos_}')
    
print('\n')
print('Detailed Grammar Category (POS tag):')
for token in doc[:9]:
    print(f'    {token.text} --> {token.tag_}')
    
print('\n')
print('Named Entity (NER):')
for token in doc[:9]:
    print(f'    {token.text} --> {token.ent_type_}')
        
print('\n')
print('Role in the Sentence (Dependency Parse):')
for token in doc[:9]:
    print(f'    {token.text} --> {token.dep_} --> {token.head.text}')
````

### Dependancy Parsing
##### Which word do I rely on grammatically?
> Note: In most sentences the verb is the semantic core

Ex: The quick brown fox jumps over the lazy dog
```/
          jumps
         /     \
      fox       over
      /   \        |
   The  quick    dog
         |
      brown

```\
fox   → nsubj → jumps
dog   → pobj  → over
over  → prep  → jumps
quick → amod  → fox
brown → amod  → fox
The   → det   → fox`

Some meanings:
```\
nsubj → subject
pobj  → object
prep  → preposition
amod  → adjective
det   → determinant

````python
for token in doc:
    print(token.text, token.dep_, token.head.text, token.head.pos_,
            [child for child in token.children])
````
