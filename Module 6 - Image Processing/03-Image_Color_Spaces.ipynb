### Split a Color Image into channels

```python
cv2.split(img_rgb)
```

### Images generated by each channel

```python
# Crea una figura con subgráficos (1 fila, 3 columnas)
fig, axs = plt.subplots(2, 3, figsize=(13, 8))  # Ajusta el tamaño con figsize

r, g, b = cv2.split(img_rgb)

# Crea canales vacíos (ceros) del mismo tamaño
zeros = np.zeros_like(r)

# Reconstruye imágenes RGB con un solo canal activo
r_img = cv2.merge([r, zeros, zeros])  # (R, 0, 0)
g_img = cv2.merge([zeros, g, zeros])  # (0, G, 0)
b_img = cv2.merge([zeros, zeros, b])  # (B, 0, B)

cmaps = ['Reds', 'Greens', 'Blues']
images = [r_img, g_img, b_img]
#-----------------------------------

for i, (cmap, img) in enumerate(zip(cmaps, images)):
    axs[0][i].imshow(img, cmap=cmap)
    axs[0][i].set_title("Red Scale")
    axs[0][i].axis('off')   # Removes axis numbers form the plot.

    axs[1][i].imshow(img_rgb)
    axs[1][i].set_title("Original")
    axs[1][i].axis('off')

# Ajusta el espaciado y muestra
plt.tight_layout()
plt.show()
```

### Algunos color maps...

```python
''' Muestra algunos colormaps. 'viridis' es la escala por defecto '''

colormaps = ['viridis', 'plasma', 'hsv', 'cividis', 'Reds', 'Greens', 'Blues', 'Greys']
n = len(colormaps)

fig, axs = plt.subplots(2, 4, figsize=(n * 2, 3))
for n, cmap in enumerate(colormaps):
    row, col = divmod(n, 4)  # Es equivalente a usar la división entera // y el operador módulo % por separado:
    axs[row][col].imshow([range(0, 50, 1)], aspect='auto', cmap=cmap)
    axs[row][col].set_title(cmap)
    axs[row][col].axis('off')

plt.show()
```

### Convert the original image into L*a*b Color Space

```python
labSpace = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2HSV)
labSpace = cv2.cvtColor(labSpace, cv2.COLOR_BGR2RGB)
print("Image shape:", labSpace.shape)
labSpace
```

#### ¿Qué es el espacio de color Lab?
El espacio de color Lab (o Lab*) es un modelo de color diseñado para aproximarse a cómo los humanos perciben los colores. A diferencia de RGB o BGR, que dependen de mezclas de luz (rojo, verde, azul), Lab separa la luminosidad del color:

- L: Luminosidad (lightness), va de 0 (negro) a 100 (blanco).
- a: Representa el eje verde-rojo (valores negativos son verdes, positivos son rojos).
- b: Representa el eje azul-amarillo (valores negativos son azules, positivos son amarillos).

##### ¿Por qué no se ve "correcta" la imagen directamente con imshow( )?
Si intentas mostrar img_lab directamente con plt.imshow(img_lab), los colores se verán raros porque matplotlib interpreta los valores de L, a, b como si fueran R, G, B, lo cual no es correcto. Lab no es un espacio de color para visualización directa en la mayoría de las herramientas; es más útil para procesamiento.

```python
# Separa los canales L, a, b
l, a, b = cv2.split(labSpace)

# Crea subgráficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Canal L (luminosidad)
axs[0].imshow(l, cmap='gray')
axs[0].set_title("L (Luminosidad)")
axs[0].axis('off')

# Canal a (verde-rojo)
axs[1].imshow(a, cmap='gray')
axs[1].set_title("a (Verde-Rojo)")
axs[1].axis('off')

# Canal b (azul-amarillo)
axs[2].imshow(b, cmap='gray')
axs[2].set_title("b (Azul-Amarillo)")
axs[2].axis('off')

plt.tight_layout()
plt.show()
```

#### Notas
Canales en gris: Usamos cmap='gray' porque cada canal (L, a, b) es una matriz de un solo valor por píxel. No hay un cmap estándar que represente directamente el eje a (verde-rojo) o b (azul-amarillo) en su forma perceptual.
Interpretación:
- L parece una versión en blanco y negro de la imagen.
- a muestra áreas verdes (valores bajos) y rojas (valores altos).
- b muestra áreas azules (valores bajos) y amarillas (valores altos).

### Convert the original image to the HSV (Hue, Saturation, Value)

```python
saturation = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2HSV)
saturation = cv2.cvtColor(saturation, cv2.COLOR_BGR2RGB)
saturation
```

##### ¿Qué es el espacio de color HSV?
El espacio de color HSV (Hue, Saturation, Value) representa los colores de una manera más alineada con la percepción humana:

- H (Hue): Tono, el "color puro" (rojo, verde, azul, etc.), va de 0 a 179 en OpenCV (la mitad de 360° porque usa una escala comprimida).
- S (Saturation): Saturación, la intensidad o pureza del color, va de 0 (gris) a 255 (color pleno).
- V (Value): Valor o brillo, la intensidad de luz, va de 0 (negro) a 255 (blanco).
##### Ventajas
- Más fácil para detectar colores específicos (por ejemplo, todos los tonos de rojo) porque el tono (H) está separado del - brillo (V) y la saturación (S).
- Útil en tareas como detección de objetos por color o ajustes de iluminación.

```python
# Separa los canales H, S, V
h, s, v = cv2.split(saturation)

# Crea subgráficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Canal H (Tono)
axs[0].imshow(h, cmap='hsv')  # 'hsv' para una representación de tonos
axs[0].set_title("H (Tono)")
axs[0].axis('off')

# Canal S (Saturación)
axs[1].imshow(s, cmap='gray')
axs[1].set_title("S (Saturación)")
axs[1].axis('off')

# Canal V (Valor)
axs[2].imshow(v, cmap='gray')
axs[2].set_title("V (Valor)")
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
